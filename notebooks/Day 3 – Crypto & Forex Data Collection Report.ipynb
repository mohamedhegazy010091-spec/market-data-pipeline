```python
import pandas as pd
import yfinance as yf
import os
from datetime import datetime , timedelta
```

```python
crypto_symbols = {
    "BTC-USD":"crypto",
    "ETH-USD":"crypto"
}

forex_symbols = {
    "EURUSD=X":"forex",
    "USDJPY=X":"forex"

}
```

```python
end_date = datetime.today()
start_date = end_date - timedelta(days=365*2)
interval = '1d'
```

```python
file_name = "Day_3_Crypto_Forex_Data"
os.makedirs(file_name , exist_ok = True)

def upload_data(symbol , asset_type):
    df =yf.download(symbol , start = start_date, end = end_date, interval = interval)
    df.columns = df.columns.get_level_values(0)
    df.reset_index(inplace = True)
    df = df[['Date','Open','High','Low','Close','Volume']]
    df['Symbol'] = symbol
    df['Asset_Type'] = asset_type
    df.dropna(inplace=True)
    return df
```

```python
crypto_data  = []
for symbol , asset_type in crypto_symbols.items():
    df = upload_data(symbol , asset_type)
    crypto_data.append(df)
    df.to_csv(f"{file_name}/{symbol.replace('-','_')}.csv",index=False)
```

```python
forex_data = []
for symbol , asset_type in forex_symbols.items():
    df = upload_data(symbol , asset_type)
    forex_data.append(df)
    df.to_csv(f"{file_name}/{symbol.replace('=X','')}.csv",index=False)
    
```

```python
btc = pd.read_csv(f"{file_name}/BTC_USD.csv")
eth = pd.read_csv(f"{file_name}/ETH_USD.csv")
eurusd = pd.read_csv(f"{file_name}/EURUSD.csv")
usdjpy = pd.read_csv(f"{file_name}/USDJPY.csv")
all_data = pd.concat(
    [btc, eth, eurusd, usdjpy],
    ignore_index=True
)
all_data.to_csv(
    f"{file_name}/All_Crypto_Forex_Data.csv",
    index=False
)
```
