```python
import pandas as pd
import yfinance as yf
```

```python
banknifty = yf.download("^NSEBANK",period='5y', interval='1d')
```

```python
banknifty
```

```python
banknifty = banknifty.reset_index()
banknifty['Range'] = banknifty['High'] - banknifty['Low']
```

```python
banknifty
```

```python
banknifty = banknifty[['Date', 'Open', 'High', 'Low', 'Close', 'Range']]
```

```python
banknifty
```

```python
sensex = yf.download("^BSESN", period="5y", interval="1d")
sensex = sensex.reset_index()
sensex['Range'] = sensex['High'] - sensex['Low']
sensex = sensex[['Date', 'Open', 'High', 'Low', 'Close', 'Range']]
sensex
```

```python
banknifty = banknifty.tail(1000)
sensex = sensex.tail(1000)
```

```python
banknifty.to_csv("BankNifty_1000_Sessions.xlsx", index=False)
sensex.to_csv("Sensex_1000_Sessions.xlsx", index=False)
```
