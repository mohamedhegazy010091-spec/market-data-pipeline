```python
import os
import pandas as pd
```

```python
df = pd.read_csv("Day_2_Stock_Data/All_Stocks_Data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns = [col.strip().title().replace(" ", "_") for col in df.columns]
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```python
df.sort_values(by=['Date', 'Symbol'], inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
invalid_High = df[
    (df['High'] < df['Low']) |
    (df['High'] < df['Open']) |
    (df['High'] < df['Close'])
]
invalid_High.shape
```

```python
invalid_Low = df[
    (df["Low"] > df["High"]) |
    (df["Low"] > df["Open"]) |
    (df["Low"] > df["Close"])
]
invalid_Low.shape   
```

```python
invalid_Low.empty & invalid_High.empty
```

```python
df = df[df["Volume"]>= 0]
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
cleaned_data = os.makedirs("Day_4_Cleaned_Data", exist_ok=True)
```

```python
df.to_csv("Day_4_Cleaned_Data/Cleaned_Stock_Data.csv", index=False)
```
